```
import numpy as np
%matplotlib inline
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```
figsize(12.5,5)
gamma=stats.gamma
parameters= [(1,0.5),(9,2),(3,0.5),(7,0.5)]
x=np.linspace(0.001,20,150)
for alpha,beta in parameters:
  y=gamma.pdf(x,alpha,scale=1./beta)
  lines=plt.plot(x,y,label="(%0.1f,%0.1f)"%(alpha,beta),lw=3)
  plt.fill_between(x,y,0,alpha=0.2,color=lines[0].get_color())
  plt.autoscale(tight=True)
plt.legend()
plt.title("Gamama distribution for the different value of alpha and beta");
```

```
figsize(12.5,5)
params=[(2,5),(1,1),(0.5,0.5),(5,5),(20,4),(5,1)]
x=np.linspace(0.01,.99,100)
beta=stats.beta
for a,b in params:
  y=beta.pdf(x,a,b)
  lines=plt.plot(x,y,label="(%.1f,%.1f)"%(a,b),lw=3)
  plt.fill_between(x,0,y,alpha=0.2,color=lines[0].get_color())
plt.ylim(0)
plt.legend(loc="upper left",title="(a,b)-parameters")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("The beta distribution for the different values of alpha and beta")
```
